```python
# Total Transaction_Amount per customer_id

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Loaded variable 'df' from URI: c:\Users\HimakarRaju\Desktop\PreSkilling\Python\Data_Science_And_Visualization\Milestone1\customer_transactions.csv
df = pd.read_csv(r'c:\Users\HimakarRaju\Desktop\PreSkilling\Python\Data_Science_And_Visualization\Milestone1\customer_transactions.csv')


def Transaction_Amount_per_customer_id(df):
    # Performed 1 aggregation grouped on column: 'Customer_ID'
    df = df.groupby(['Customer_ID']).agg(Transaction_Amount_sum=('Transaction_Amount', 'sum')).reset_index()
    return df

df_clean1 = Transaction_Amount_per_customer_id(df.copy())
print(df_clean1)

x = df_clean1['Customer_ID']
y = df_clean1['Transaction_Amount_sum']

# Plotting the data
plt.bar(x, y)
plt.xlabel('Customer ID')  # Corrected label for x-axis
plt.ylabel('Total Transaction Amount')  # Corrected label for y-axis
plt.xticks(x, rotation=90)  # Rotating x-axis labels for better visibility
plt.title('Total Transaction Amount per Customer ID')  # Adding a title to the plot
plt.show()
```

```python

import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

file = './customer_transactions.csv'

df = pd.read_csv(file)



data = df

# 1. Descriptive Statistics: You can get an overview of your dataset using:
print(df.describe())


# 2. Total Transaction Amount by City: You can calculate and visualize the total transaction amounts for each city:
city_totals = data.groupby(
    'City')['Transaction_Amount'].sum().sort_values(ascending=False)

city_totals.plot(kind='bar', title='Total Transaction Amount by City')
plt.show()

# Transaction Amount per Customer : You can calculate and visualize the total transaction amounts for each customer:
df_Trans_Amt_Per_ID = df.groupby(['Customer_ID']).agg(
    Transaction_Amount_sum=('Transaction_Amount', 'sum')).reset_index()
df_Trans_Amt_Per_ID.plot(kind='line', title='Transaction Amount per Customer')
plt.show()

# 3. Transaction Type Distribution: Visualize the proportion of different transaction types:
sns.countplot(x='Transaction_Type', data=data)
plt.show()

sns.boxplot(x='Account_Type', y='Transaction_Amount', data=data,)
plt.show()


# 4. Transaction Amount Distribution: Plot the distribution of transaction amounts to identify skewness or outliers:
data['Transaction_Amount'].plot(
    kind='hist', bins=30, title='Transaction Amount Distribution')
plt.show()
```
